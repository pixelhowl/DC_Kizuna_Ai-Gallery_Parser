```python
import os
import pandas as pd
from datetime import datetime
gall = 'kizunaai';year=datetime.now().year; month =datetime.now().month-1 # 갤러리, 연, 월1
baseloc=os.path.abspath('./%s/%s/%s' % (gall,year,month))
idf = pd.read_json(baseloc+'post.json')

output_filename='sum_ai.csv'
cur_data = pd.read_csv(os.path.join(baseloc, output_filename), index_col=0)

noholo_filename='sum_noholo.csv'
noholo_data = pd.read_csv(os.path.join(baseloc, noholo_filename), index_col=0)

old_month = month - 1
old_year = year
if month == 1:
    old_month = 12
    old_year = year - 1
oldloc=os.path.abspath('./%s/%s/%s' % (gall,old_year,old_month))
old_data = pd.read_csv(os.path.join(oldloc, output_filename), index_col=0)
```

```python
def gen_barlength(data):
    standard = data.loc[1,"총 언급 횟수"]
    barlength = 18.56 / standard
    barlengths = [18.56]
    for i in range(2,11):
        out = cur_data.loc[i,"총 언급 횟수"]*barlength
        barlengths.append(out)
#         print(f"{i}번째 바 길이: {out}")
    return barlengths
```

```python
cur_barlengths=gen_barlength(cur_data)
noholo_barlengths = gen_barlength(noholo_data)
```

```python
import pptx_utils as utils
from pptx import Presentation
from pptx.chart.data import ChartData
import pandas as pd
import os
import re
from datetime import datetime
from pptx.util import Cm, Inches, Pt

prs = Presentation("template.pptx")


month_pattern = r"[0-9]+월"
count_pattern = "("
increase_pattern = r"▼|▲|■|NEW"

def update_slide(slide_num, data, barlengths):
    textboxes = []
    pictures = []
    rectangles = []

    for shape in prs.slides[slide_num].shapes:
        print(f"{shape.name} leftx: {shape.left}, topy:{shape.top}, width:{shape.width}, height:{shape.height}")
        if shape.name.find("직사각형") != -1:
            rectangles.append(shape)
        elif shape.name.find("TextBox") != -1:
            textboxes.append(shape)
        elif shape.name.find("그림") != -1 or shape.name.find("Picture") != -1:
            pictures.append(shape)

    textboxes.sort(key=lambda x: x.top, reverse=False)
    pictures.sort(key=lambda x: x.top, reverse=False)
    rectangles.sort(key=lambda x: x.top, reverse=False)

    print("text------")
    count_idx = 0
    for shape in textboxes:
        whole_text = "".join([r.text for r in shape.text_frame.paragraphs[0].runs])
        if slide_num == 0 and whole_text == "69":
            target_text = str(data.index[data["Vtuber"] == "키즈나 아이"].tolist()[0])
            print(target_text)
            text_frame = shape.text_frame
            color = (240,240,240)
            p = text_frame.paragraphs[0]
            utils.replace_text_only(p, target_text, color)
            
        elif re.search(month_pattern, whole_text) is not None: 
            target_text = whole_text.replace(str(old_month), str(month))
            color = (240,240,240)
            text_frame = shape.text_frame
            p = text_frame.paragraphs[0]
            utils.replace_text_only(p, target_text, color)
        elif whole_text.find(count_pattern) != -1:
            if slide_num == 0:
                detail_count_text = data[data["Vtuber"] == "키즈나 아이"]["단어(언급 수)"].values[0]
                print(detail_count_text)
            else:
                detail_count_text = data.loc[count_idx+1, "단어(언급 수)"]
                count_idx += 1
            text_frame = shape.text_frame
            p = text_frame.paragraphs[0]
            if len(detail_count_text) > 55:
                detail_count_text = detail_count_text[:54]+"..."
            utils.replace_text_only(p, detail_count_text, color)
        print(f"{shape.name} leftx: {shape.left}, topy:{shape.top}, width:{shape.width}, height:{shape.height}, text: {whole_text}")

    print("pic------")
    for shape in pictures: 
        print(f"{shape.name} leftx: {shape.left}, topy:{shape.top}, width:{shape.width}, height:{shape.height}")
        left = shape.left
        top = shape.top
        width = shape.width
        height = shape.height
        # Replace image:shape.width = barlengths[count_idx]
        if slide_num == 0 and shape.name == "그림 8":
            utils.replace_picture(prs.slides[slide_num], shape, f"{baseloc}/wordcloud.png")

    count_idx = 0
    print("shape------")
    for shape in rectangles:
        whole_text = "".join([r.text for r in shape.text_frame.paragraphs[0].runs])
        if  re.search(increase_pattern, whole_text) is not None:
            if slide_num == 0:
                count = data[data["Vtuber"] == "키즈나 아이"]["총 언급 횟수"].values[0]
                increase_count = data[data["Vtuber"] == "키즈나 아이"]["전월 대비 순위"].values[0]
            else:
                count = data.loc[count_idx+1, "총 언급 횟수"]
                increase_count = data.loc[count_idx+1, "전월 대비 순위"]
                shape.width = Cm(barlengths[count_idx])
                count_idx += 1

            if increase_count.find("▼") != -1:
                increase_color = (0,113,197)
            elif increase_count.find("▲") != -1 or increase_count.find("NEW") != -1:
                increase_color = (192,0,0)
            elif increase_count.find("■") != -1:
                increase_color = (13,13,13)
            else:
                increase_color = (13,13,13)
            
            target_texts = [str(count)+" ", str(increase_count)]
            colors = [(13,13,13), increase_color]

            text_frame = shape.text_frame
            p = text_frame.paragraphs[0]
            utils.replace_texts_with_diff_color(p, target_texts, colors)
        print(f"{shape.name} leftx: {shape.left}, topy:{shape.top}, width:{shape.width}, height:{shape.height}, text: {whole_text}")


```
